 Import Required Libraries


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

Load Dataset

```python
df=pd.read_csv(r"C:\Users\mohas\Desktop\financial_consumer\Financial Consumer Complaints.csv")
```

# Wrangling Data

 Initial Data Inspection

```python


df.head()
df.info()
```

 Drop Irrelevant  Columns

```python
df=df.drop(columns=['ZIP code','Consumer consent provided?','Tags'])
```

 Clean Column Names to Standard Snake Case

```python

df.columns=df.columns.str.strip().str.replace('-','_').str.replace('?',"").str.replace(' ',"_").str.lower()
```

```python
df=df.rename(columns={"date_sumbited":"date_submitted"})
```

 Convert Date Columns to Datetime

```python
df['date_received']=pd.to_datetime(df['date_received'],format='%m/%d/%y')
```

```python
df['date_submitted']=pd.to_datetime(df['date_submitted'],format='%m/%d/%y')
```

```python
df.columns=df.columns.str.strip().str.replace('-','_').str.replace('?',"").str.replace(' ',"_").str.lower()
```

 Standardize Missing & Misformatted Text Field

```python

df['sub_product']=df['sub_product'].replace('""','Un known')
```

```python
df['sub_issue']=df['sub_issue'].replace('""','Un known')
```

Manage missing company public response

```python

df['company_public_response']=df['company_public_response'].fillna(
  'Pending to provide a public response')
```

 Drop Rows Where sub_issue is Missing

```python
df=df.dropna(subset=['sub_issue'])
```

Fill Missing Categorical Values with Mode

```python
most_frequent_state = df['state'].mode()[0]
```

```python
df['state']=df['state'].fillna(most_frequent_state)
```

```python
most_frequent_disputed=df['consumer_disputed'].mode()[0]
```

```python
df['consumer_disputed']=df['consumer_disputed'].fillna(most_frequent_disputed)
```
